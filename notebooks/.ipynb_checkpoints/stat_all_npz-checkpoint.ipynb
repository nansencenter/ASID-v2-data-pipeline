```python
import glob
import argparse
import datetime
import os
from os.path import basename, dirname, join
from operator import itemgetter
import random
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
concentration_scale = [0, 1, 2, 10, 20, 30, 40, 50, 60, 70, 80, 90, 91, 92]
sod_scale = [0, 80, 81, 82, 83, 84, 85, 86, 87, 88, 89, 91, 93, 95, 96, 97, 98]
form_scale = [i for i in range (22)]+[99]
list_combinations =[]
for sod in sod_scale :
    for form in form_scale :
        name_comb = str(sod)+'_'+str(form)
        list_combinations.append(name_comb)
```

```python
npz_files = sorted(glob.glob('/Data/analysis/output_summary/*all_npz.npz'))
```

```python
def creation_matrix_combi_con(npz_files, list_combination, concentration_scale, dist_thresholds, visualization=False):    
    """
    This function creates a graph. 
    It represents the frequency of each combination according to concentration.
    In addition, if the distance to the zone is too small (below the threshold)
    then the data from the zone will not be taken into account.
    
    Parameters
    ----------
    npz_files : list
        all npz files in the folder.
    list_combination : list
        list of the all possible combinations.
    concentration_scale : list
        list that contains all the levels of concentration according to ASID-v2 manual.
    dist_thresholds : float
        minimum border distance threshold.
    visualization : boolean
        if it makes graphiques.
    
    Returns
    -------
    result : list
        list of combinations and their frequencies for each concentration.
    """
    #initialization
    x_length = len(list_combination)
    y_length = len(concentration_scale)
    sol = np.zeros((y_length, x_length))
    list_logical_graph=[True]*len(list_combinations)
    list_index_combination_present = []
    list_combination_present = []
    
    #creation of the matrix
    for npz_file in npz_files:
        list_output = np.load(npz_file)['array']
        for output in list_output :
            if (output[10] > dist_thresholds) : # if the distance of the output is below the threshold, we don't keep
                if output[0] == 2: #total concentration == 02 (open water)
                    combi = "0_0"
                    sol[0][0] += 1
                    list_logical_graph[0]=False
                    if 0 not in list_index_combination_present :
                        list_index_combination_present.append(0)  
                for ice in range(3): # in a output there are 3 data for the 3 most present ice
                    if output[1+ice*3]==(-9): 
                        continue
                    if output[2+ice*3]==(-9): 
                        continue
                    if output[3+ice*3]==(-9): 
                        continue
                    if output[3+ice*3]==(99): 
                        continue
                    combi = str(int(output[2+ice*3])) + '_' + str(int(output[3+ice*3]))
                    index_con = concentration_scale.index(int(output[1+ice*3]))
                    index_comb = list_combination.index(combi)
                    sol[index_con][index_comb] += 1
                    list_logical_graph[index_comb] = False
                    if index_comb not in list_index_combination_present :
                        list_index_combination_present.append(index_comb)
    
    
    #combination present
    list_ind_comb_present_sort = sorted(list_index_combination_present)
    for ind in list_ind_comb_present_sort :
        list_combination_present.append(list_combination[ind])
    print(list_combination_present)
    
    
    #Cleanning
        # matrix cleanning
    sol1 = sol.compress(np.logical_not(list_logical_graph), axis =1)
    sol1 = np.asarray(sol1, dtype = int)
        # cleanning of the return 
    sol_return =[]
    c=0
    for combi in list_combination_present :
        sol_return.append([combi, sol1[:,c]])
        c += 1
    
    # visualization
    if visualization :
        plt.figure(figsize=(5,15))
        sol1 = sol1.astype(float)
        sol1[sol1==0] = np.nan
        plt.imshow(sol1.T, clim=[0, 10000])
        plt.colorbar(shrink=0.1)
        plt.title('Frequency of combinations depending on concentration')
        plt.yticks(range(len(list_combination_present)), list_combination_present)
        plt.ylabel('Combinations (Sod_form)')
        plt.xlabel('Concentration')
        plt.show()
    return sol_return
```

```python
array = creation_matrix_combi_con(npz_files, list_combinations, concentration_scale, 12.8, visualization=True)
for i in array :
    print(i)
```

```python
def analysis(combi_array, cut_threshold = 0.66, visualization=False):
    """
    This function allows to generate a vector with the expected combinations 
    according to the analysis of the data made by the function creation_matrix_combi_con.
    Two conditions :
    -the combination must be present in at least three ice concentrations
    -only 2/3 (or cut_threshold) of the most frequent combinations are kept.
    
    Parameters
    ----------
    combi_array : array
        array obtained by the function creation_matrix_combi_con.
    cut_threshold : float 
        percentage of combinations we want to keep between 0 and 1.
    visualization : boolean
        if it makes graphiques.
    
    Returns
    -------
    result : list
        list of combinations we will keep during preprocessing
    """
    
    list_without_1_2_con_sum=[]
    #open water
    open_w_sum =['0_0',combi_array[0][1][0]]

    for combi in combi_array:
        #we sort the vector and if the 3rd value is equal to 0 then we delete it
        frequence = sorted(combi[1], reverse=True)
        if frequence[2] == 0:
            continue
        freq_sum = sum(frequence)
        list_without_1_2_con_sum.append([combi[0],freq_sum])
    #we keep only the 2/3 most present
    list_without_1_2_con_sum.append(open_w_sum)# we add open water
    list_comb_sum_sort = sorted(list_without_1_2_con_sum, key=itemgetter(1), reverse=True)
    #labels, values = *zip(*list_comb_sum_sort)
    labels, values = np.array(list_comb_sum_sort).T
    values = list(map(float, values))
    if visualization:
        plt.bar(labels, values)
        plt.title('Frequency of combinations')
        plt.xlabel('Combinations (sod_form)')
        plt.ylabel('Frequency')
        plt.show()
        
    lenght = len(list_comb_sum_sort)
    index_cut = int(cut_threshold * lenght)
    list_final_sum = list_comb_sum_sort[:index_cut]
    list_final=[]
    for comb in list_final_sum:
        list_final.append(comb[0])
    
    return list_final
```

```python
vector_combination = analysis(array, cut_threshold = 0.66, visualization=True)
print(vector_combination)
print(len(vector_combination))
```
